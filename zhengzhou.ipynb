```python
import pandas as pd
import datetime  
import numpy as np
from LunarSolarConverter.LunarSolarConverter import LunarSolarConverter, Lunar, Solar
from matplotlib import pyplot as plt
import seaborn as sns
import pylab
%matplotlib inline
from sklearn.model_selection import train_test_split

np.random.seed(7)
ctis = pd.read_csv('/mnt/d/ctis.csv', sep='|')
converter = LunarSolarConverter()
ctis['date'] = pd.to_datetime(ctis['tran_date'], format='%Y%m%d', errors='ignore')
ctis['day'] = ctis['date'].dt.day
ctis['month'] = ctis['date'].dt.month
ctis['weekday'] = ctis['date'].dt.dayofweek
ctis['Solar'] = ctis['date'].map(lambda date: Solar(date.year, date.month, date.day))
ctis['Lunar'] = ctis['Solar'].map(lambda date: converter.SolarToLunar(date))
ctis['LunarDay'] = ctis['Lunar'].map(lambda date: date.lunarDay)
ctis['LunarMonth'] = ctis['Lunar'].map(lambda date: date.lunarMonth)
ctis['day1_amt'] = ctis['amt'].shift(1)
ctis['day2_amt'] = ctis['amt'].shift(2)
```

```python
from sklearn.preprocessing import normalize
from sklearn.model_selection import GridSearchCV

from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, lasso_path, enet_path
```

```python
ctis = ctis.dropna()
normed = normalize(ctis[['day', 'month', 'weekday', 'LunarDay', 'LunarMonth', 'day1_amt']].values, norm='l2')
ctis['day'] = normed[:,0]
ctis['month'] = normed[:,1]
ctis['weekday'] = normed[:,2]
ctis['LunarDay'] = normed[:,3]
ctis['LunarMonth'] = normed[:,4]
ctis['day1_amt'] = normed[:,5]
```

```python
data_2017 = ctis[ctis['date'] < pd.to_datetime('20180101', format='%Y%m%d', errors='ignore')]
data_2018 =  ctis[ctis['date'] >= pd.to_datetime('20180101', format='%Y%m%d', errors='ignore')]
X_train = data_2017[['day', 'month', 'weekday', 'LunarDay', 'LunarMonth', 'day1_amt']].values
y_train = data_2017['amt'].values
X_test = data_2018[['day', 'month', 'weekday', 'LunarDay', 'LunarMonth', 'day1_amt']].values
y_test = data_2018['amt'].values
```
